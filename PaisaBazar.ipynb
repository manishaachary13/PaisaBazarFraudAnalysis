# **EDA(Exploratory Data Analysis)**

#### 1.

```python
# ---- Outlier Detection ----
def plot_boxplots(columns):
    plt.figure(figsize=(15, 6))
    for i, col in enumerate(columns):
        plt.subplot(1, len(columns), i + 1)
        sns.boxplot(y=paisa_df[col])
        plt.title(f"Boxplot of {col}")
    plt.show()

# Select numerical columns for outlier detection
num_cols = ['Annual_Income', 'Outstanding_Debt', 'Credit_Utilization_Ratio', 'Monthly_Balance']
plot_boxplots(num_cols)
```

1. **Why did you choose this chart?**
* Boxplots are used to detect outliers and understand the spread of numerical data. Since financial data like income, debt, and credit utilization often contain extreme values, a boxplot helps visualize these variations effectively.

2. **Purpose of the chart?**
* The purpose is to analyze the distribution of key financial metrics such as Annual Income, Outstanding Debt, Credit Utilization Ratio, and Monthly Balance, identifying any potential outliers that could indicate anomalies or fraud.

3. **Insights:**
* Annual Income & Outstanding Debt: Both show significant outliers, meaning some individuals have extremely high incomes or debts compared to the majority.
* Credit Utilization Ratio: Some customers have a very high ratio, indicating over-reliance on credit, which can lead to poor credit scores.
* Monthly Balance: A few customers maintain a very high balance, suggesting they might be better at managing finances, or they could be high-income individuals.

4. **Business Recommendations:**
* For Risk Assessment: Customers with extremely high outstanding debt and credit utilization ratios should be flagged for deeper risk analysis before loan approvals.
* For Fraud Detection: Investigate customers with exceptionally high income or balance to ensure the data is accurate and not fabricated.
* For Credit Policy Improvement: Offer financial guidance to customers with high credit utilization to help them manage their credit better and improve their financial health.

```python
# ---- Data Distributions ----
def plot_distributions(columns):
    plt.figure(figsize=(15, 6))
    for i, col in enumerate(columns):
        plt.subplot(1, len(columns), i + 1)
        sns.histplot(paisa_df[col], kde=True, bins=30)
        plt.title(f"Distribution of {col}")
    plt.show()

plot_distributions(num_cols)
```


**1. Why did you choose this chart?**  
Histograms with KDE plots are used to understand the distribution of numerical variables. This helps in identifying skewness, multimodal patterns, and common value ranges, which are crucial for financial analysis.

**2. Purpose of the chart?**  
The purpose is to examine how **Annual Income, Outstanding Debt, Credit Utilization Ratio, and Monthly Balance** are distributed among customers. Understanding these distributions helps assess financial stability and risk.

**3. Insights:**  
- **Annual Income:** Positively skewed (right-tailed), meaning most customers earn lower salaries, with a few high-income earners.  
- **Outstanding Debt:** Bimodal distribution, indicating two distinct groups‚Äîone with low debt and another with significantly high debt levels.  
- **Credit Utilization Ratio:** Normally distributed, with most customers falling within the range of 25-40%.  
- **Monthly Balance:** Right-skewed, meaning most people have a lower balance, with a few maintaining significantly high balances.  

**4. Business Recommendations:**  
- **For Loan Approvals:** Since most customers have lower incomes and high debt, lending policies should incorporate stricter risk assessment criteria.  
- **For Debt Management:** Offer financial advisory services to help customers maintain a **healthy debt-to-income ratio** and avoid financial stress.  
- **For Credit Utilization:** Encourage customers to keep their utilization ratio below **30%** to improve credit scores and borrowing capacity.
